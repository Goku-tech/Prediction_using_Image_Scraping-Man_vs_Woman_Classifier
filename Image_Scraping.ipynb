# **Using Image Data Generator to convert the scraped images into train data and labels:**

Image Data Generator plays a very important role in augmenting images which is a method of preventing overfitting by rotating or flipping or shifting or zooming or changing the brightness of the original image by creating a copy without affecting the original image. 

It also increases the size of the training dataset.

Thus it plays a very important role in allowing the model to generalize.


```
train_datagen=ImageDataGenerator(rescale=1/255.0,horizontal_flip=True,vertical_flip=True,rotation_range=30,fill_mode='nearest')
```

```
train_data=train_datagen.flow_from_directory("/content/drive/My Drive/Artificial Intelligence/AI Projects/Prediction using Scraping/Man_vs_Woman/Train_set",target_size=(28,28),batch_size=40,class_mode='binary')
```

# **Using Image Data Generator to convert the scraped images into validation data and labels:**

```
validation_datagen=ImageDataGenerator(rescale=1/255.0)
```

```
valid_data=validation_datagen.flow_from_directory("/content/drive/My Drive/Artificial Intelligence/AI Projects/Prediction using Scraping/Man_vs_Woman/Validation_set",batch_size=12,class_mode='binary',target_size=(28,28))
```

## **CNN Model(Using Functional API):**

```
#Input Layer:
input_layer=tf.keras.layers.Input((28,28,3))
#Convolutional Layer:
conv1_layer=tf.keras.layers.Conv2D(64,(3,3),activation='relu')(input_layer)
#MaxPooling Layer:
maxpool1_layer=tf.keras.layers.MaxPool2D((2,2))(conv1_layer)
#Covolutional Layer:
conv2_layer=tf.keras.layers.Conv2D(32,(3,3),activation='relu')(maxpool1_layer)
#MaxPool Layer:
maxpool2_layer=tf.keras.layers.MaxPool2D((2,2))(conv2_layer)
#Flattened Layer:
flatten_layer=tf.keras.layers.Flatten()(maxpool2_layer)
#Hidden Layer:
hidden_layer1=tf.keras.layers.Dense(units=512,activation='relu')(flatten_layer)
#Hidden Layer:
hidden_layer2=tf.keras.layers.Dense(units=256,activation='relu')(hidden_layer1)
#Output Layer:
output_layer=tf.keras.layers.Dense(units=1,activation='sigmoid')(hidden_layer2)

#Model:
model=tf.keras.Model(inputs=[input_layer],outputs=output_layer)

```

```
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
#Callback:Early Stopping
callback=tf.keras.callbacks.EarlyStopping(monitor='accuracy',baseline=0.4,mode='max',patience=3)

analysis=model.fit(train_data,steps_per_epoch=50,epochs=10,callbacks=[callback],validation_data=valid_data)
```
